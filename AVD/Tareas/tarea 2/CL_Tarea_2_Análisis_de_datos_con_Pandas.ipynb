## **8. Cuente el numero de jovenes, adultos y ancianos**
---
Un aspecto interesante a analizar es cómo la edad influye en el diagnóstico de la diabetes. Para ello, añadiremos una columna llamada `ciclo de vida`, que clasificará a los pacientes en los siguientes grupos:


- `joven`: edad <26

- `adulto`: edad >= 26 y edad < 59

- `anciano`: edad >= 59

**Instrucciones**

1. Agregue la columna `ciclo de vida` al conjunto de datos con la clasificación correspondiente.
2. Cuente la cantidad de pacientes en cada categoría y almacene el resultado en la variable `conteo_edad`
3. Para evitar modificaciones en el conjunto de datos original, cree una copia llamada `data_1` y realice los cambios sobre ella.


<details>    
<summary>
    <font size="3" color="darkgreen"><b>Pistas</b></font>
</summary>

* Recuerde inicializar la columna nueva
* puede usar la asignación por medio de la indexación condicional

````python
## VARIABLE CALIFICADA: conteo_edad

### ESCRIBA SU CÓDIGO AQUÍ ### (~ 5 línea de código)
data_1 = data.copy()
data_1.loc[data_1['Age']<26,'ciclo de vida'] = 'joven'
data_1.loc[(data_1['Age']>=26) & (data_1['Age']<60),'ciclo de vida'] = 'adulto'
data_1.loc[data_1['Age']>=59,'ciclo de vida'] = 'anciano'
conteo_edad = data_1['ciclo de vida'].value_counts()
### FIN DEL CÓDIGO ###

conteo_edad
````
